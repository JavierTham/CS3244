```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import tensorflow as tf

import tensorflow_datasets as tfds
import matplotlib.pyplot as plt
import os
import cv2

train_data_path = os.path.join("data", "train")
test_data_path = os.path.join("data", "test")

train_dataset = tf.keras.utils.image_dataset_from_directory(
    train_data_path,
    label_mode = "categorical",
    seed = 0,
    #batch_size = 64
)

test_dataset = tf.keras.utils.image_dataset_from_directory(
    test_data_path,
    label_mode = "categorical",
    seed = 0,
    #batch_size = 64
)

train_dataset = train_dataset.unbatch()
test_dataset = test_dataset.unbatch()

def np_conv(dataset):
    return [(example.numpy(), label.numpy()) for example, label in dataset]

test_np = np_conv(test_dataset)
train_np = np_conv(train_dataset)
```

```python
X_train = [example for example, label in train_np]
y_train = [label for example, label in train_np]
X_test = [example for example, label in test_np]
y_test = [label for example, label in test_np]
```

```python
X_train = np.concatenate(X_train)
X_test = np.concatenate(X_test)
y_test = np.asarray(y_test)
y_train = np.asarray(y_train)
y_test = np.argmax(y_test, axis=1)
y_train = np.argmax(y_train, axis=1)
```

```python
X_train128 = []
X_test128 = []

for i in os.listdir('data/test'):
    for j in os.listdir('data/test/' + i):
        tmp = cv2.imread('data/test/' + i + '/' + j)
        print(i+j)
        tmp = cv2.resize(tmp, (128,128))
        X_test128.append(tmp)
        #Visualize a random one
        if(i+j == "8_Negative99.png"):
            plt.imshow(cv2.cvtColor(tmp, cv2.COLOR_BGR2RGB))
            plt.show()
            
X_test = np.asarray(X_test128)            
            
for i in os.listdir('data/train'):
    for j in os.listdir('data/train/' + i):
        tmp = cv2.imread('data/train/' + i + '/' + j)
        tmp = cv2.resize(tmp, (128,128))
        X_train128.append(tmp)        

X_train = np.asarray(X_train128) 
```

```python
#Edge Detection
#Auto-canny function, source: https://pyimagesearch.com/2015/04/06/zero-parameter-automatic-canny-edge-detection-with-python-and-opencv/

def auto_canny(image, sigma=0.33):
    # Compute the median of the single channel pixel intensities
    v = np.median(image)

    # Apply automatic Canny edge detection using the computed median
    lower = int(max(0, (1.0 - sigma) * v))
    upper = int(min(255, (1.0 + sigma) * v))
    return cv2.Canny(image, lower, upper)
```

```python
X_train_canny = []
X_test_canny = []

for i in os.listdir('data/test'):
    for j in os.listdir('data/test/' + i):
        tmp = cv2.imread('data/test/' + i + '/' + j)
        print(i+j)
        tmp = cv2.resize(tmp, (128,128))
        tmp = auto_canny(tmp)
        X_test_canny.append(tmp)
        #Visualize a random one
        if(i+j == "8_Negative99.png"):
            plt.imshow(cv2.cvtColor(tmp, cv2.COLOR_BGR2RGB))
            plt.show()
            
X_test_edge = np.asarray(X_test_canny)            
            
for i in os.listdir('data/train'):
    for j in os.listdir('data/train/' + i):
        tmp = cv2.imread('data/train/' + i + '/' + j)
        tmp = cv2.resize(tmp, (128,128))
        tmp = auto_canny(tmp)
        X_train_canny.append(tmp)        

X_train_edge = np.asarray(X_train_canny) 
    
```

```python
X_train_ef = np.asarray(list(map(lambda x: x.flatten(), X_train_edge)))
X_test_ef = np.asarray(list(map(lambda x: x.flatten(), X_test_edge)))
normalize(X_train_ef,copy=False)
normalize(X_test_ef,copy=False)
```

```python
ens_edge = RandomForestClassifier(n_jobs = -1)
result_ens_edge = cross_val_score(ens_edge , X_train_ef, y_train, cv = kf, scoring = 'f1_micro', n_jobs = -1)
```

```python
print('(W)F1 Score: %.3f' % f1_score(y_test, y_pred, average='weighted'))
print('(W)Recall: %.3f' % recall_score(y_test, y_pred, average='weighted'))
print('(W)Precision: %.3f' % precision_score(y_test, y_pred, average='weighted'))
print('(M)F1 Score: %.3f' % f1_score(y_test, y_pred, average='macro'))
print('(M)Recall: %.3f' % recall_score(y_test, y_pred, average='macro'))
print('(M)Precision: %.3f' % precision_score(y_test, y_pred, average='macro'))
print('RF + Canny accuracy for training set: %f' % ens_edge.score(X_train_ef, y_train))
print('RF + Canny accuracy for test set: %f' % ens_edge.score(X_test_ef, y_test))
```
